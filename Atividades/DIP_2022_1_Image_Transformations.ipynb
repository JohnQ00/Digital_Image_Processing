#### (1) Crop and flip an image using Numpy array indexing.


```python
image = cv.imread('musashi.jpg')
cv2_imshow(image)
```

#### (2) Implement image translation using Numpy and OpenCV.


```python
cv2_imshow(image)
```

```python
height, width = image.shape[:2]
shift = 50

# shift image
rolled = np.roll(image, shift, axis=[0, 1])
# black out shifted parts
rolled = cv.rectangle(rolled, (0, 0), (width, shift), 0, -1)
rolled = cv.rectangle(rolled, (0, 0), (shift, height), 0, -1)
cv2_imshow(rolled)
```

```python
new_height, new_width = height / 8, width / 8
T = np.float32([[1, 0, new_width], [0, 1, new_height]])
image_translated = cv.warpAffine(image, T, (width, height))
cv2_imshow(image_translated)
```

#### (3) Implement image rotation using Numpy and OpenCV.


```python
cv2_imshow(image)
```

```python
image_rotated = np.rot90(image, 3)
cv2_imshow(image_rotated)
```

```python
image_rotated = cv.rotate(image, cv.ROTATE_90_CLOCKWISE)
cv2_imshow(image_rotated)
```

#### (4) Implement image resizing using OpenCV.


```python
height, width = image.shape[:2]
new_size = (height//2, width//2)
image_resized = cv.resize(image, new_size)
cv2_imshow(image_resized)
```

#### (5) Implement bitwise operations: AND, OR, XOR.


```python
rectangle = cv.rectangle(np.zeros((300, 300), dtype="uint8"), (20, 20), (280, 280), 255, -1)
cv2_imshow(rectangle)

circle = cv.circle(np.zeros((300,300),dtype='uint8'), (150, 150), 150, 255, -1)
cv2_imshow(circle)
```

```python
bitwise_and = cv.bitwise_and(rectangle, circle)
cv2_imshow(bitwise_and)
```

```python
bitwise_or = cv.bitwise_or(rectangle, circle)
cv2_imshow(bitwise_or)
```

```python
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv2_imshow(bitwise_or)
```

#### (6) Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.

```python
f = image
cv2_imshow(f)
```

```python
g = np.zeros(f.shape[:2], dtype="uint8")
cv.rectangle(g, (75, 75), (500, 550), 255, -1)
cv2_imshow(g)
```

```python
h = cv.bitwise_and(image, image, mask=g)
cv2_imshow(h)
```
